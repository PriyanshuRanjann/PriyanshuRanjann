```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
plt.plot([1,2,3],[4,5,6],[3,6,7])
plt.show()
```

```python
plt.plot([1,2,3],[4,5,6])

plt.title('Ranjans plot')
plt.xlabel("This is abscissa")
plt.ylabel("This is ordinate")
plt.show()
```

```python
style.use('ggplot')
a=[2,4,56]
b=[12,13,14]
c=[6,3,4]
d=[12,15,35]
plt.plot(a,b,label="FIRST",Linewidth=5)
plt.plot(c,d,label="SECOND")
plt.legend()
plt.show()
```

```python
plt.pie(a,)
plt.legend()
plt.show()
```
